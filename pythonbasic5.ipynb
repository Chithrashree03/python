# Data Structure

## List

#### Create a list of 5 cities. Add a new city to the end.

```python
cities=["Hassan","Mysore","Bangalore","Chikmagalur","Mangalore"]
print(cities)
cities.append("shivamogga")
print(cities)
```

#### Replace the second city with "Pune".¶

```python
cities[1]="pune"
print(cities)
```

####  Print all cities starting from the third one.

```python
print(cities[3:])   
```

#### Count how many cities are in the list.

```python
print(len(cities))
```

## Tuple

#### Create a tuple of 3 programming languages.

```python
language = ("python", "java", "sql")
print(language)
```

#### Check if "Python" is in the tuple.

```python
if "python" in language:
 print("true")
else:
 print("false")
```

#### Print each language using a loop.

```python
language = ("python", "java", "sql")
for i in language:
    print(i)
```

#### Convert the tuple to a list and add "JavaScript".

```python
language = ["python", "java", "sql"]
language.append("javascript")
print(language)
```

## Dictionary

#### Create a dictionary with product name, price, and quantity.

```python
Milk = {"name":"Nandini","price":"30rs","quantity":"1lt"}
print(Milk)
```

#### Update the quantity to 5

```python
Milk["quantity"]="5lt"
print(Milk)
```

####  Print only the keys in the dictionary.

```python
print(Milk.keys())
```

#### Use a loop to print each key and value.

```python
for key,value in Milk.items():
    print(f"key:{key}, Value:{value}")
```
