## User defined functions:

###  Write a function to check if a number is even or odd.

```python
def evenorodd(num):
    
  if num %2==0:
     return "even"
  else:
     return "odd"
```

```python
evenorodd(20)
```

### Create a function that returns the factorial of a number.

```python
def factorial(n):
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers.")
    if n == 0:
        return 1
    else:
        n>=1
        return n*factorial(n-1)
```

```python
print(factorial(10))
```

### Write a function to find the maximum of three numbers.

```python
def maximum(num1,num2,num3):
    if num1>=num2 and num1>=num3:
        return num1
    elif num2>=num1 and num2 >= num3:
        return num2
    else:
        return num3
```

```python
maximum(40,20,10)
```

## Lambda Function:

### Use a lambda to multiply two numbers.

```python
multiply =lambda a,b:a*b
print(multiply(8,5))
```

### Use filter() to get even number from a list.

```python
a=[1,2,3,5,6,7,8,9,10,12,13,15,18]
even=list(filter(lambda a:a%2==0,a))
print(even)
```

### Use map() to covert a list of strings to uppercase.

```python
str=["hello","madam"]
uppercase=list(map(lambda string:string.upper(),str))
print(uppercase)
```
